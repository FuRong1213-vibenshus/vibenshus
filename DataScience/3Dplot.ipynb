```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
fig = plt.figure(figsize=(14,7))
ax = plt.axes(projection='3d')
theta = np.linspace(0, 2 * np.pi, 30)
w = np.linspace(0, 0.5*np.pi, 50)
w, theta = np.meshgrid(w, theta)

r = np.sin(w)
Y = r*np.cos(theta)
Z = r*np.sin(theta)

ax = plt.axes(projection='3d')
ax.plot_surface(w, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface');
ax.view_init(25, -75)
```
